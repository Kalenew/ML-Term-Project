```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from sklearn import preprocessing,metrics
```

```python
credit = fetch_openml(name = 'credit-approval')
X = credit.data
y = credit.target 
loc = np.where(np.isnan(X[:,0]))
X[:] = np.nan_to_num(X[:])
# X[:,1] = np.nan_to_num(X[:,1])
X = StandardScaler().fit_transform(X)
X_train, X_test, y_train, y_test = \
        train_test_split(X, y, test_size=.2,random_state=42)

le = preprocessing.LabelEncoder()
cytrain=le.fit(y_train)
cytrain=le.transform(y_train)
cytest= le.transform(y_test)
```

```python

# # Generate dummy data
# x_train = np.random.random((1000, 20))
# y_train = np.random.randint(2, size=(1000, 1))
# x_test = np.random.random((100, 20))
# y_test = np.random.randint(2, size=(100, 1))

model = Sequential()
model.add(Dense(64, input_dim=len(X[1]), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
print(model.summary())
model.compile(loss='mean_squared_error',
              optimizer='SGD',
              metrics=['accuracy'])
history = model.fit(X_train, cytrain, validation_data=(X_test, cytest),batch_size=200,
          epochs=2000)
score = model.evaluate(X_test, cytest, batch_size=200)


# model2 = Sequential()
# model2.add(Dense(64, input_dim=len(X[1]), activation='relu'))
# model2.add(Dropout(0.5))
# model2.add(Dense(128, activation='relu'))
# model2.add(Dropout(0.5))
# model2.add(Dense(64, activation='relu'))
# model2.add(Dropout(0.5))
# model2.add(Dense(1, activation='sigmoid'))
# print(model2.summary())

# model2.compile(loss='mean_squared_error',
#               optimizer='RMSprop',
#               metrics=['accuracy'])
# history2 = model2.fit(X_train, cytrain, validation_split=0.2,batch_size=200,
#           epochs=2000)
# score2 = model2.evaluate(X_test, cytest, batch_size=200)

# model3 = Sequential()
# model3.add(Dense(64, input_dim=len(X[1]), activation='relu'))
# model3.add(Dropout(0.5))
# model3.add(Dense(128, activation='relu'))
# model3.add(Dropout(0.5))
# model3.add(Dense(64, activation='relu'))
# model3.add(Dropout(0.5))
# model3.add(Dense(1, activation='sigmoid'))
# print(model3.summary())
# model3.compile(loss='mean_squared_error',
#               optimizer='adam',
#               metrics=['accuracy'])
# history3 = model3.fit(X_train, cytrain, validation_split=0.2,batch_size=200,
#           epochs=2000)
# score3 = model3.evaluate(X_test, cytest, batch_size=200)


# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
# plt.plot(history2.history['accuracy'])
# plt.plot(history2.history['val_accuracy'])
# plt.plot(history3.history['accuracy'])
# plt.plot(history3.history['val_accuracy'])
plt.title('accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='lower right')
# plt.legend(['SGD_train', 'SGD_test','RMSprop_train2', 'RMSprop_test2','adam_train3', 'adam_test3'], loc='lower right')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
# plt.plot(history2.history['loss'])
# plt.plot(history2.history['val_loss'])
# plt.plot(history3.history['loss'])
# plt.plot(history3.history['val_loss'])
plt.title('loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')
# plt.legend(['SGD_train', 'SGD_test','RMSprop_train2', 'RMSprop_test2','adam_train3', 'adam_test3'], loc='upper right')
plt.show()
```
