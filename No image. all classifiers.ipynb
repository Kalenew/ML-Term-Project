```python
# imports
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing,metrics
from sklearn.datasets import fetch_openml
from sklearn.model_selection import KFold,StratifiedKFold
from sklearn.ensemble import AdaBoostClassifier

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier,BernoulliRBM

from sklearn.metrics import precision_recall_fscore_support,precision_score,recall_score,f1_score,precision_recall_curve
```

```python
# Classifiers
names = [#"Nearest Neighbors", 
         "Linear SVM", 
         "Decision Tree", "Random Forest","AdaBoost", "MLP",
#          "BernoulliRBM"
        ]

classifiers = [
    #KNeighborsClassifier(3),
    SVC(kernel ="linear"),
    SVC(kernel ="rbf"),
    SVC(kernel ="poly"),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5,n_estimators=200, max_features=1),
#     AdaBoostClassifier(DecisionTreeClassifier(max_depth=5),n_estimators=200),
#     MLPClassifier(alpha=1,max_iter=1000),
#     BernoulliRBM(n_components=60)
]
```

```python
credit = fetch_openml(name = 'credit-approval')
X = credit.data
y = credit.target 
loc = np.where(np.isnan(X[:,0]))
X[:] = np.nan_to_num(X[:])
```

```python
i=0
X = StandardScaler().fit_transform(X)
# X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size = .2,random_state =42)
kf = KFold(n_splits=10)
kf = StratifiedKFold(n_splits=10)
StratifiedKFold
#kf.get_n_splits(X)

# precisionK=[]
# precisionC=[]
# recallK=[]
# recallC=[]
# fK=[]
# fC=[]
# suptK=[]
# suptC=[]
scoreCat1=[]
scoreCat2=[]
scoreCat3=[]
scoreCat4=[]
scoreCat5=[]
scoreCat1r=[]
scoreCat2r=[]
scoreCat3r=[]
scoreCat4r=[]
scoreCat5r=[]

scoreCat1a=[]
scoreCat2a=[]
scoreCat3a=[]
scoreCat4a=[]
scoreCat5a=[]

scoreCat1f=[]
scoreCat2f=[]
scoreCat3f=[]
scoreCat4f=[]
scoreCat5f=[]
for trn_idx,tst_idx in kf.split(X,y):
    X_train, X_test = X[trn_idx],X[tst_idx]
    y_train, y_test = y[trn_idx],y[tst_idx]
    
    le = preprocessing.LabelEncoder()
    cytrain=le.fit(y_train)
    cytrain=le.transform(y_train)
    cytest= le.transform(y_test)
    print("==================Fold  %s ===================\n" % i )  
    
    j=0
    for name, clf in zip(names, classifiers):
            score_array=[]
            clf.fit(X_train, cytrain)
            #score = clf.score(X_test, cytest)
            
            decision = clf.predict(X_test)
            #decision= le.transform(np.rint(decision))
            if j==0:
                scoreCat1.append(precision_score(cytest, decision,average='weighted'))
                scoreCat1r.append(recall_score(cytest, decision,average='weighted'))
                scoreCat1a.append(clf.score(X_test, cytest))
                scoreCat1f.append(f1_score(cytest, decision,average='weighted'))
                precision_recall_curve(cytest, decision)
                #scoreCat1.append(precision_recall_fscore_support(cytest, decision, average='weighted'))
            elif j==1:
                scoreCat2.append(precision_score(cytest, decision,average='weighted'))
                scoreCat2r.append(recall_score(cytest, decision,average='weighted'))
                scoreCat2a.append(clf.score(X_test, cytest))
                scoreCat2f.append(f1_score(cytest, decision,average='weighted'))
                #scoreCat2.append(precision_recall_fscore_support(cytest, decision, average='weighted'))  
            elif j==2:
                scoreCat3.append(precision_score(cytest, decision,average='weighted'))
                scoreCat3r.append(recall_score(cytest, decision,average='weighted'))
                scoreCat3a.append(clf.score(X_test, cytest))
                scoreCat3f.append(f1_score(cytest, decision,average='weighted'))
                #scoreCat3.append(precision_recall_fscore_support(cytest, decision, average='weighted'))  
            elif j==3:
                scoreCat4.append(precision_score(cytest, decision,average='weighted'))
                scoreCat4r.append(recall_score(cytest, decision,average='weighted'))
                scoreCat4a.append(clf.score(X_test, cytest))
                scoreCat4f.append(f1_score(cytest, decision,average='weighted'))
                #scoreCat4.append(precision_recall_fscore_support(cytest, decision, average='weighted'))  
            elif j==4: 
                scoreCat5.append(precision_score(cytest, decision,average='weighted'))
                scoreCat5r.append(recall_score(cytest, decision,average='weighted'))
                scoreCat5a.append(clf.score(X_test, cytest))
                scoreCat5f.append(f1_score(cytest, decision,average='weighted'))
                #scoreCat5.append(precision_recall_fscore_support(cytest, decision, average='weighted'))
            j+=1
            print("Classification report for classifier :\n%s\n"
                      % (metrics.classification_report(cytest, decision)))
    i+=1   
avgPcs1=np.mean(scoreCat1)
avgPcs2=np.mean(scoreCat2)
avgPcs3=np.mean(scoreCat3)
avgPcs4=np.mean(scoreCat4)
avgPcs5=np.mean(scoreCat5)

avgRcl1=np.mean(scoreCat1r)
avgRcl2=np.mean(scoreCat2r)
avgRcl3=np.mean(scoreCat3r)
avgRcl4=np.mean(scoreCat4r)
avgRcl5=np.mean(scoreCat5r)
(avgPcs1,avgPcs2,avgPcs3,avgPcs4,avgPcs5)
```
